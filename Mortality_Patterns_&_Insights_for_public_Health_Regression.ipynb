#**3.Data Cleaning**

Data cleaning was performed to enhance the dataset's quality and consistency. Two key steps were taken:

Column Removal: Since this data is of USA the columns named 'Jurisdiction.of.Occurrence' was redundant (all values were same). similarly, 'AnalysisDate' column was not adding value to the data. Hence they were subsequently deleted from the dataset. This streamlines the dataset and removes unnecessary information.
Gender Standardization: In the 'Sex' column, entries for gender were not uniform, with variations like 'F,' 'Female,' 'Male,' and 'M.' To ensure consistency and uniformity, the mutate
 function was employed to standardize the gender entries, ensuring that all gender values follow a consistent format.

Deletion of 2 columns :

```r
#since the columns named 'Jurisdiction.of.Occurrence' & 'AnalysisDate' does not add any value to the data, we can delete them
data <- data[, !colnames(data) %in% c("Jurisdiction.of.Occurrence","AnalysisDate")]
head(data)
```

Since in the column "Sex", the entries of gender are not uniform (F,Female,Male,M). To bring uniformity we use Mutate() function as :

```r
# Replace 'F' with 'Female' and 'M' with 'Male' in the 'sex' column
data <- data %>%
  mutate(Sex = case_when(
    Sex == 'F' ~ 'Female',
    Sex == 'M' ~ 'Male',
    TRUE ~ Sex
  ))

head(data,10)
```

##**Comparing Models**

To compare the five simple linear regression models and determine which one is the best choice, let's examine their key characteristics:

**Model 1 (Diseases of the Heart):**
- R-squared: 0.1296
- Residual standard error: 1330
- Coefficient for AgeNumeric: 178.658

**Model 2 (Cerebrovascular Diseases):**
- R-squared: 0.1314
- Residual standard error: 326.3
- Coefficient for AgeNumeric: 44.177

**Model 3 (Symptoms, Signs, and Abnormal Clinical Findings):**
- R-squared: 0.03411
- Residual standard error: 89.46
- Coefficient for AgeNumeric: 5.8515

**Model 4 (Diabetes Mellitus):**
- R-squared: 0.1632
- Residual standard error: 138.3
- Coefficient for AgeNumeric: 21.2620

**Model 5 (Influenza and Pneumonia):**
- R-squared: 0.1259
- Residual standard error: 94.96
- Coefficient for AgeNumeric: 12.5414

To analyse the models -

- **R-squared** measures how well the model explains the variance in the data. Higher values are better, indicating a better fit to the data.
- **Residual standard error** measures the average deviation between the predicted values and the actual values. Lower values are better, indicating a better fit to the data.
- **Coefficient for AgeNumeric** indicates the strength and direction of the relationship between age and the health issue. A larger coefficient suggests a stronger relationship.

- **Observation:** **Model 4 (Diabetes Mellitus)** has the highest R-squared value (0.1632), indicating that it explains the most variance in the data, and the lowest residual standard error (138.3), meaning it provides a better fit to the data.
- The coefficient for AgeNumeric in Model 4 (21.2620) is relatively high, suggesting a strong relationship between age and diabetes mellitus.

Therefore, based on these metrics, **Model 4 (Diabetes Mellitus)** appears to be the best choice for analyzing the relationship between age and health issues among the five models. It indicates that age is a significant predictor of diabetes mellitus.
